#Import libraries and load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset (replace with your file path)
df = pd.read_csv("Titanic-Dataset.csv")

# Quick peek at data
print(df.shape)
print(df.head())
```

```python
#Generate summary statistics
```

```python
# General info
print(df.info())
print(df.describe(include="all"))  # all includes categorical

# Missing values
print(df.isnull().sum())
```

```python
#Univariate Analysis (Distributions of Features)
#Histograms
```

```python
df.hist(figsize=(12, 8), bins=30)
plt.show()
```

```python
#Boxplots
```

```python
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
#Bivariate Analysis (Relationships between Features)
#Correlation Heatmap
```

```python
plt.figure(figsize=(10, 6))

# Select only numeric columns for correlation
corr = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

```python
#Pairplot (for numeric features)
```

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.show()
```

```python
#Scatterplot
```

```python
print(df.columns)
```

```python
sns.scatterplot(data=df, x="SibSp", y="Parch", hue="Survived")
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(df, x="Age", color="Survived", marginal="box")
fig.show()
```

```python
fig = px.histogram(df, x="Fare", color="Pclass", marginal="violin")
fig.show()
```

```python
fig = px.scatter(df, x="Age", y="Fare", color="Sex", size="SibSp")
fig.show()
```
